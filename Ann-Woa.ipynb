```python
 #Test for hanoi benchmark
networks_dataframe_benchHanoi=pd.read_csv('C:/systembenchHanoi.csv')
networks_dataframe_xgboost_Hanoi=networks_dataframe_benchHanoi[['EDGE_Headloss_network','NodeToEdge_Node_DCIN2',
                     'EDGE_Velocity_network','NodeToEdge_Node_DCOUT2','EDGE_Length_network',
                     'EDGE_Friction_network','NodeToEdge_NODE_AVE_DIM','NodeToEdge_NODE_MINpathRHW',
                     'EDGE_BetweennessCentrality_UndirectedRHW','NodeToEdge_Node_DCOUT1','NodeToEdge_Node_DCIN1',
                     'NodeToEdge_Demand_network','EDGE_BetweennessCentrality_directededlength']]
networks_dataframe_xgboost_Hanoi['EDGE_Diameter_network']=networks_dataframe_benchHanoi['EDGE_Diameter_network']
```

```python
# Remove Outlier from dataset
def remove_outliers(networks_dataframe):    
    for i in networks_dataframe.columns:
        if (networks_dataframe[i].dtypes == int) | (networks_dataframe[i].dtypes == float):
            Lower_outliers = networks_dataframe[i].mean() - 4.2*networks_dataframe[i].std()
            Upper_outliers = networks_dataframe[i].mean() + 4.2*networks_dataframe[i].std()
            networks_dataframe = networks_dataframe[(networks_dataframe[i]<=Upper_outliers) & (networks_dataframe[i]>=Lower_outliers)]
    return networks_dataframe
dataset_xgboost = remove_outliers(networks_dataframe_xgboost)
```

```python
frames_xg_bench = [dataset_xgboost,networks_dataframe_xgboost_Hanoi]
result_combine_xghanoi = pd.concat(frames_xg_bench)
```

```python
# Scaled Data
scaler_MinMaxScaler = MinMaxScaler()
num_vars_MinMaxScalerXg =result_combine_xghanoi.columns
result_combine_xghanoi[num_vars_MinMaxScalerXg] = scaler_MinMaxScaler.fit_transform(result_combine_xghanoi[num_vars_MinMaxScalerXg])
```

```python
train_val_xghanoi=result_combine_xghanoi.iloc[:len(dataset_xgboost),:]
X2_train_xghanoi = train_val_xghanoi.iloc[:,0:-1]
y2_train_xghanoi = train_val_xghanoi.iloc[:,-1]

test_val_xghanoi=result_combine_xghanoi.iloc[len(dataset_xgboost):,:]
test_data_xghanoi = test_val_xghanoi.iloc[:,0:-1]
test_label_xghanoi = test_val_xghanoi.iloc[:,-1]
```

```python
#Using WOA for ANN hyperparameter tuning

X_train, y_train, X_test, y_test = X2_train_xghanoi , y2_train_xghanoi, test_data_xghanoi , test_label_xghanoi
# Define your fitness function
def fitness(solution):
    n_hidden = int(solution[0])
    alpha = float(solution[1])
    lr = float(solution[2])

    model = MLPRegressor(
        hidden_layer_sizes=(n_hidden, n_hidden // 2),
        alpha=alpha,
        learning_rate_init=lr,
        max_iter=1000,
        early_stopping=True,
        activation='relu',
        solver='adam',
        random_state=42
    )

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    return mse

# Define the problem
problem = {
    "obj_func": fitness,
    "bounds": FloatVar(lb=[2, 1e-7, 1e-5], ub=[100, 1e-1, 1e-1]),
    "minmax": "min",
}

# Initialize and run the optimizer
model = OriginalWOA(epoch=100, pop_size=30)#epoch=100, pop_size=50
agent = model.solve(problem)
best_solution = agent.solution
best_fitness = agent.target
print("Best hyperparameters (neurons, alpha, lr):", best_solution)
print(f"Neurons: {int(best_solution[0])}")
print(f"alpha: {int(best_solution[1])}")
print(f"Learning rate: {best_solution[2]}")
print(f"MSE: {best_fitness}")
```

```python
#Use best hyperparameters in final ANN model
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error ,r2_score
n_hidden , alpha , lr = list(best_solution)
n_hidden=int(n_hidden)
best_model = MLPRegressor(hidden_layer_sizes=(n_hidden,n_hidden//2),
                          alpha=alpha,
                          learning_rate_init=lr,
                          max_iter=2500,
                          activation='relu',
                          solver='adam',
                          random_state=42)
best_model.fit(X_train, y_train)

y_pred = best_model.predict(X_test)

from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Test MSE: {mse:.4f}")
print(f"Test RÂ² score: {r2:.4f}")

result_ANN_Pred_withoutoversamp=pd.DataFrame({'y_predANN_xghanoi':y_pred})
result_ANN_Pred_withoutoversamp.to_csv('ANN_Pred_withoutoversamp.csv', index=False)

r2scoreANN_xghanoi = r2_score(test_label_xghanoi, y_pred)
maeANN_xghanoi = mean_absolute_error(test_label_xghanoi, y_pred) 
mseANN_xghanoi = mean_squared_error(test_label_xghanoi, y_pred) #default=True 
rmseANN_xghanoi = np.sqrt(mseANN_xghanoi)

print('r2scoreANN_xghanoi',r2scoreANN_xghanoi)
print('maeANN_xghanoi',maeANN_xghanoi)
print('mseANN_xghanoi',mseANN_xghanoi)
print('rmseANN_xghanoi',rmseANN_xghanoi)
```
